```python
import nltk
from nltk.corpus import brown
```

```python
#using the words from corpus.brown
words = brown.words(categories= 'news')
```

```python
freqDist = {}
setOfWords = list(set(words))
```

```python
for word in words:
    if word not in freqDist:
        freqDist[word] = 1
    else:
        freqDist[word] += 1
```

```python
freqDist = sorted(freqDist.items(), key=lambda x: x[1], reverse=True)
```
